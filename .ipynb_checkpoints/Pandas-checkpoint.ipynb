## Pandas Tutorial

### Load data into Pandas

```python
import pandas as pd

df = pd.read_csv('pokemon_data.csv')
```

```python
print(df)
```

```python
# prints the top 3 columns, use df.tail() to print bottom rows
print(df.head(3))
```

### Reading Data in Pandas

```python
# Printing Headers
print(df.columns)
```

```python
# Printing each Column

print(df['Name'])

# The same can be written as print(df.Name)
```

```python
# Prints specific columns

print(df['Name'][0:5])
```

```python
# Printing multiple columns

print(df[['Name', 'Type 1', 'HP']])
```

```python
# Print each row

print(df.iloc[1])
```

```python
# Print multiple rows

print(df.iloc[1:4])
```

```python
# Read a specific location

print(df.iloc[1,1])
```

```python
# Iterating through data

for index, row in df.iterrows():
    print(index,row.Name)
```
